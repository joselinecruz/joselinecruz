```python
#Reto: Análisis de deserción de empleados con sus sistemas operativos utilizados
import pandas as pd
file_id = '1PaxlDPlPynxNaXtty55zy9ZdbtTvuWrx'
url = f'https://drive.google.com/uc?id={file_id}'
Test_csv = pd.read_csv(url)

file_id = '1Yzwg4jF4JJTChSpwaFmWE3tAedNYUbGX'
url = f'https://drive.google.com/uc?id={file_id}'
Test2_csv = pd.read_csv(url)

file_id = '1bsUijbZkjuJPSYjRjMopXeafTAVDZfmL'
url = f'https://drive.google.com/uc?id={file_id}'
Train_csv = pd.read_csv(url)

file_id = '1EDC3eWE6sTxZIdwWHDTwO-9wsTfHhzbd'
url = f'https://drive.google.com/uc?id={file_id}'
Train2_csv = pd.read_csv(url)
#Unir los dataframe
df_train = pd.merge(Train_csv, Train2_csv, on="Employee_ID", how="left")
df_test  = pd.merge(Test_csv,  Test2_csv,  on="Employee_ID", how="left")
df_total = pd.concat([df_train, df_test], axis=0, ignore_index=True)
#5
df_total[" Computer_OS"] = df_total[" Computer_OS"].fillna("no registrado")
porcentaje_computadora = (
    df_total[" Computer_OS"]
    .value_counts(normalize=True) * 100
).reset_index()
#print(df_total.columns)
porcentaje_computadora.columns = ["Sistema_Operativo", "Porcentaje"]
print(porcentaje_computadora)
#El porcentaje de personas que usan iOS, Android o "No registrado" como su sistema operativo en su dispositivo móvil. Se esperan 3 renglones en esta respuesta.
df_total[" Mobile_OS"] = df_total[" Mobile_OS"].fillna("no registrado")

porcentaje_movil = (
    df_total[" Mobile_OS"]
    .value_counts(normalize=True) * 100
).reset_index()

porcentaje_movil.columns = ["Mobile_OS", "Porcentaje"]
print(porcentaje_movil)
#La edad promedio de la gente que utiliza cada sistema operativo de escritorio.
porcentaje_Edad = (df_total.groupby(" Computer_OS")["Age"].mean())
#El nivel educativo promedio de la gente que utiliza cada sistema operativo de escritorio.
porcentaje_Educacion = (df_total.groupby(" Computer_OS")["Education_Level"].mean())
#El número total de tickets generados por la gente, con cada sistema operativo de escritorio.
total_ticket = (df_total.groupby(" Computer_OS")[[" Computer_tickets", " Mobile_tickets"]].sum())
```

```python
with pd.ExcelWriter("DS_C2_SC3_JoselineCruzHernandez.xlsx", engine="openpyxl") as writer:
    porcentaje_Edad.to_excel(writer, sheet_name="porcentaje_Edad", index=False)
    porcentaje_Educacion.to_excel(writer, sheet_name="porcentaje_Educacion", index=False)
    porcentaje_movil.to_excel(writer, sheet_name="porcentaje_movil", index=False)
    total_ticket.to_excel(writer, sheet_name="total_ticket", index=False)
    porcentaje_movil.to_excel(writer, sheet_name="porcentaje_movil", index=False)
    porcentaje_computadora.to_excel(writer, sheet_name="porcentaje_computadora", index=False)
from google.colab import files
files.download("DS_C2_SC3_JoselineCruzHernandez.xlsx")
print("Archivo DS_C2_SC3_JoselineCruzHernandez.xlsx creado correctamente.")
```
