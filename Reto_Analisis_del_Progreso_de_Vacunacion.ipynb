```python
# A) Extraer la informacion del archivo.
import pandas as pd
file_id = '1rl-Q-zK76HeW7ADphMleeyqsAfndv12o'
url = f'https://drive.google.com/uc?id={file_id}'
cou_vaccination = pd.read_csv(url)
print(cou_vaccination.head ())

# B) Mostrar la estructura y tipos de datos de cada columna para identificar que operaciones puedes realizar, que las columnas son fechas tipo datetime64.
print("B")
print("Estructura del DataFrame:")
print(cou_vaccination.info())
print(cou_vaccination["date"].dtype)

# C) Cantidad de Vacunas aplicadas de cada compañia (con base en como lo reporta cada pais en la columna vacciones, en otras palabras agrupe por vacciones y realice la sumatoria).
#Separamos las vacunas
vacunas_expandidas = cou_vaccination['vaccines'].str.split(',').explode().str.strip()
#Contar cuántas veces aparece cada una
conteo_vacunas = vacunas_expandidas.value_counts()
#Mostrar el resultado
print("C")
print(conteo_vacunas)

# D) Determinar la cantidad de vacunas aplicadas en todo el mundo.
total_mundo = cou_vaccination['total_vaccinations'].sum()
print("D")
print("Total de vacunas aplicadas del mundo:")
print(total_mundo)

# E) Calcular el promedio de vacunas aplicadas por pais.
prom_por_pais = cou_vaccination.groupby('country')['total_vaccinations'].mean()
print("E")
print("Promedio total por pais:")
print(prom_por_pais)

# F) Determinar la cantidad de vacuas aplicadas el dia 29/01/21
cou_vaccination['date'] = pd.to_datetime(cou_vaccination['date'],dayfirst=True)
suma_por_fecha = cou_vaccination.groupby('date')['total_vaccinations'].sum()
print("F")
print("Vacunas el 29/01/2021:")
print(suma_por_fecha.loc['2021-01-29'])

```

```python
# G) Crear dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicacion entre las columnas daily_vaccionations y daily_vaccionations_row.
conDiferencias = cou_vaccination.copy()
conDiferencias["diferencias"] = (conDiferencias["daily_vaccinations"].fillna(0) - conDiferencias["daily_vaccinations_raw"].fillna(0))
print("G")
print(conDiferencias.head())
```

```python
# H) Obtener el periodo de tiempo entre el registro con fecha mas reciente y el registro con fecha mas antigua.
cou_vaccination["date"] = pd.to_datetime(cou_vaccination["date"], dayfirst=True)
# Obtener la fecha más antigua y más reciente
fecha_min = cou_vaccination["date"].min()
fecha_max = cou_vaccination["date"].max()
# Calcular el periodo de tiempo
periodo = fecha_max - fecha_min
print("H")
print("Fecha más antigua:", fecha_min)
print("Fecha más reciente:", fecha_max)
print("Periodo total:", periodo)
```

```python
# I) Crear un dataframe nuevo denominado conCantidad que obtenga los datos originales y una columna derivada (conVac) con la cantidad de vacunas utilizadas cada dia.
conCantidad = cou_vaccination.copy()
# Crear la columna conVac contando cuántas vacunas hay por día
conCantidad["conVac"] = conCantidad["vaccines"].str.split(", ").apply(len)
conCantidad.head()
print("I")
print(conCantidad[["date", "country", "conVac"]].head(10))

# J) Generar un datafram denominado antes20 con todos los registros que se hayan realizado antes de 20 de Diciembre 2020.
cou_vaccination["date"] = pd.to_datetime(cou_vaccination["date"])
fecha_limite = pd.to_datetime("2020-12-20")
antes20 = cou_vaccination[cou_vaccination["date"] < fecha_limite]
print ("J")
print(antes20.head(10))
```

```python
# K) Obtener un dataframe denominada pfizer con todos los registros donde se haya utilizado la vacuna pfizer.
pfizer = cou_vaccination[cou_vaccination["vaccines"].str.contains("Pfizer", case=False, na=False)]
print(pfizer.head(10))
# L) Almacenar los dataframe generados (conDiferencias conCantidad, antes20 y pfizer) en un archivo de Excel resultadoReto.xlsx donde cada dataframe ocupe una hoja diferente.
with pd.ExcelWriter("resultadoReto.xlsx", engine="openpyxl") as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias", index=False)
    conCantidad.to_excel(writer, sheet_name="conCantidad", index=False)
    antes20.to_excel(writer, sheet_name="antes20", index=False)
    pfizer.to_excel(writer, sheet_name="pfizer", index=False)
from google.colab import files
files.download("resultadoReto.xlsx")
print("Archivo resultadoReto.xlsx creado correctamente.")
```
